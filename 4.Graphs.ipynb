```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import pickle
import matplotlib.dates as mdates
```

```python
pkl_file = open('daily_candidate_sentiments_df.pkl', 'rb')
df = pickle.load(pkl_file)
pkl_file.close()


cols=[i for i in df.columns if i not in ["Date"]]
for col in cols:
    df[col]=pd.to_numeric(df[col])
    
df['Date'] =  pd.to_datetime(df['Date'])
df['Date'] = df['Date'].dt.date

df = df[['Date', 'Joe Biden', 'Bernie Sanders']]

df = df.melt('Date', var_name='Candidates',  value_name='Sentiment')
df = df.sort_values(by='Date')

# # g = sns.factorplot(x="Date", y="Sentiment", hue='Candidates', data=daily_candidate_sentiments_df)

palette = sns.color_palette("colorblind", 12)
# palette = sns.dark_palette("muted purple", input="xkcd")

g = sns.pointplot(x="Date", y="Sentiment", hue='Candidates', data=df)

box = g.get_position() # getting position of figure
g.set_position([box.x0, box.y0, box.width * 0.85*2.5, box.height*1.5]) # resize position

# Putting a legend to the right side
g.legend(loc='center right', bbox_to_anchor=(1.2, 0.5), ncol=1)
plt.xticks(rotation=90)
plt.grid()
plt.show(g)
```
